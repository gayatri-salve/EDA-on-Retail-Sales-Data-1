#IMPORTING THE LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#LOADING DATA
```

```python
data = pd.read_csv('D:/retail_sales_dataset.csv')
data
```

```python
#CLEANING DATA
```

```python
data.count()
```

```python
# Check for duplicates
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
print(data.head())
```

```python
#DISCRIPTIVE STATISTICS
```

```python
# Basic statistics for numeric columns

mean_sales = data['Total Amount'].mean()
median_sales = data['Total Amount'].median()
mode_sales = data['Total Amount'].mode()[0]  # Mode might have multiple values
std_dev_sales = data['Total Amount'].std()

print(f"Mean Sales: {mean_sales}")
print(f"Median Sales: {median_sales}")
print(f"Mode Sales: {mode_sales}")
print(f"Standard Deviation of Sales: {std_dev_sales}")
```

```python
#TIME SERIES ANALYSIS
```

```python
# Time Series Analysis
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='Total Amount', data=data)
plt.title('Time Series Analysis of Total Amount')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.show()
```

```python
#VISUALIZATION 
```

```python
# Customer and Product Analysis
plt.figure(figsize=(10,6))
sns.countplot(x='Product Category', data=data)
plt.title('Product Category Distribution')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()

gender_distribution = data['Gender'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=gender_distribution.index, y=gender_distribution.values)
plt.title('Gender Distribution of Customers')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(x='Quantity', y='Price per Unit', data=data)
plt.title('Quantity vs Price per Unit')
plt.xlabel('Quantity')
plt.ylabel('Price per Unit')
plt.show()

```

```python
# Heatmap
plt.figure(figsize=(10,6))
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
sns.heatmap(data[numeric_cols].corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()

# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], kde=True, bins=20)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
